```python
from datetime import datetime, timedelta
from shapely.geometry import Polygon, box
import pystac_client
import stackstac
from rasterio.enums import Resampling
import xarray as xr

def read_sentinel_image_from_stac(stac_url: str, bbox: list, start_date: datetime, end_date: datetime) -> xr.DataArray:
    """
    Read the best Sentinel-2 image from the last two weeks based on cloud coverage 
    within the San Francisco Bay Area from a STAC catalog using pystac_client.

    Parameters:
        stac_url (str): The URL of the STAC catalog.
        bbox (list): Bounding box coordinates [west, south, east, north].
        start_date (datetime): Start date for temporal filtering.
        end_date (datetime): End date for temporal filtering.

    Returns:
        xr.DataArray: The Sentinel-2 image as an xarray DataArray.

    Raises:
        ValueError: If the request to the STAC API fails or if no suitable items are found.
    """
    try:
        # Define area of interest
        area_of_interest = box(*bbox)

        # Connect to the STAC catalog
        catalog = pystac_client.Client.open(stac_url)

        # Search for Sentinel-2 items within the specified date range and AOI
        search = catalog.search(
            datetime=(start_date, end_date),
            intersects=area_of_interest,
            max_items=100,
            query={"eo:cloud_cover": {"lt": 80}},
        )
        items = search.get_all_items()

        if not items:
            raise ValueError("No suitable items found in the STAC catalog")

        # Choose the last item from the search results
        item = items[0] #[-1]
        print(item)
        
        epsg = item.properties["proj:epsg"]


        # Stack the assets (e.g., red, green, blue bands) into an xarray DataArray
        stack = stackstac.stack(
            item,
            bounds=bbox,
            snap_bounds=False,
            dtype="float32",
            epsg=4326,
            rescale=False,
            fill_value=0,
            assets=["red", "green", "blue"],
            resampling=Resampling.nearest,
            xy_coords="center",
        )

        stack = stack.compute()  # Compute lazy operations
        return stack

    except Exception as e:
        raise ValueError(f"Error accessing STAC catalog: {e}")

# Example usage:
STAC_URL = "https://earth-search.aws.element84.com/v1"  # Provided STAC URL
BBOX = [-122.51, 37.75, -122.45, 37.78] # [-123.0, 36.8, -121.5, 38.0]  # Bounding box for San Francisco Bay Area
START_DATE = datetime.utcnow() - timedelta(weeks=12)
END_DATE = datetime.utcnow()

try:
    sentinel_image = read_sentinel_image_from_stac(STAC_URL, BBOX, START_DATE, END_DATE)
    print("Sentinel-2 image loaded successfully")
    sentinel_image.sel(band=["red", "green", "blue"]).plot.imshow(
    row="time", rgb="band", vmin=0, vmax=5000)
except ValueError as ve:
    print("Error:", ve)


```

```python
import geopandas as gpd
import pystac_client
import xarray as xr
import matplotlib.pyplot as plt

# Define the area of interest (AOI) as a bounding box
aoi_bbox = (-122.55, 37.6, -122.35, 37.9)  # Example AOI around San Francisco Bay Area

# Connect to the Sentinel-2 STAC catalog
stac_url = "https://earth-search.aws.element84.com/v1"
catalog = pystac_client.Client.open(stac_url)

# Search for Sentinel-2 scenes within the AOI and specified time range
search_results = catalog.search(
    collections=["sentinel-2-l2a"],
    bbox=aoi_bbox,
    datetime="2024-01-01T00:00:00Z/2024-03-28T23:59:59Z",
    query={"eo:cloud_cover": {"lt": 20}},
)

# Get a list of item IDs from the search results
item_ids = [item.id for item in search_results.get_all_items()]

# Choose a specific scene (for demonstration purposes, we'll use the first scene in the list)
scene_id = item_ids[0]

# Access the selected scene
scene = catalog.get_item(scene_id)

# Access the assets (bands) available for the scene
assets = scene.assets

# Choose the bands of interest (e.g., blue, green, red, nir) and their corresponding asset keys
band_assets = {
    "B02": "blue",
    "B03": "green",
    "B04": "red",
    "B08": "nir",
}

# Open the selected bands using xarray and concatenate them into a single xarray dataset
datasets = []
for asset_key, band_name in band_assets.items():
    asset_href = assets[asset_key].href
    dataset = xr.open_rasterio(asset_href, chunks={"band": 1})  # Open the band raster data
    dataset = dataset.squeeze().drop("band")  # Remove the band dimension (single-band)
    dataset.name = band_name  # Assign the band name as the dataset name
    datasets.append(dataset)

# Concatenate the datasets along the band dimension
combined_dataset = xr.concat(datasets, dim="band")

# Plot RGB composite of the satellite image
rgb_image = combined_dataset.sel(band=["red", "green", "blue"]).transpose("y", "x", "band")
rgb_image.plot.imshow(robust=True, figsize=(10, 10))
plt.title("RGB Composite Sentinel-2 Image")
plt.show()

# Perform water quality analysis using the NIR band
nir_band = combined_dataset.sel(band="nir")

# Example analysis (e.g., NDVI calculation)
red_band = combined_dataset.sel(band="red")
ndvi = (nir_band - red_band) / (nir_band + red_band)
ndvi.plot.imshow(cmap="RdYlGn", robust=True, figsize=(10, 10))
plt.title("Normalized Difference Vegetation Index (NDVI)")
plt.show()
```
